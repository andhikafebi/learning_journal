# Vektor gradien (turunan parsial) masing-masing pixel

```python
R_dx,R_dy = np.gradient(R)

fig, ax = plt.subplots(ncols=3,figsize=(21,5))
ax[0].imshow(R)
ax[0].set_title('image Filter R')
ax[1].imshow(R_dx)
ax[1].set_title('dx')
ax[2].imshow(R_dy)
ax[2].set_title('dy')
```
